```python
import numpy as np
import pandas as pd
```

```python
# Data Loading and Exploration
data_path = r'D:\app\cereal\backend\data\cereal.csv'

# Reading the CSV file with the specified encoding
df = pd.read_csv(data_path, encoding='UTF-8-SIG')

# Display the first few rows of the dataset
print('First few rows of the dataset:')
display(df.head())

# Display summary of the dataset
print('\nDataset info:')
df.info()

print('\nStatistical summary:')
display(df.describe(include='all'))
```

```python
df= df.drop( ['name','mfr', 'type','sodium','fiber','carbo','potass','vitamins','shelf','weight','cups','rating','record_datetime'], axis=1)
df
```

```python
df = df.dropna()  # Drop rows with any NaN values
print('\nData after dropping NaN values:')
df
```

```python
# Splitting the above data into training and testing data
from sklearn.model_selection import train_test_split
x=df[['protein', 'fat', 'sugars']]
y=df[['calories']]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state = 9598)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# As it is the Regrssion Problem lets try all the Regression Algorith to get the best accuracy
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# To train and fit the model
model.fit(x_train,y_train)
```

```python
#Test Model
y_pred=model.predict(x_test) # Predict the target variable for the test set

print(len(y_pred)) # Show the number of predictions made
y_pred # Show predicted values
```

```python
#Evaluate Model
mape=np.mean(np.absolute(y_test - y_pred) / y_test * 100) # Calculate Mean Absolute Percentage Error (MAPE)
print('MAPE :', np.round(mape,2)) # Print Error in percentage
```

```python
#Export Model
import joblib
# Save the trained model to a file
joblib.dump(model,"cereal-calories.pkl")
```
